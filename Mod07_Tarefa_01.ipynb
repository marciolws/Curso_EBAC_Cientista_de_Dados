````python
#Foi criada uma lista com as variáveis

df_01 = df.columns.tolist()
````

````python
df_01['QTD_CATEGORIAS'] = qtd

df_01
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
#Utilizando o get_dummies nas variáveis object que não são binárias.

df_02 = pd.get_dummies(df, columns=['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
````

````python
#O dtypes mostra que foram criadas novas colunas, separando os dados das colunas objects em uint8.

df_02.dtypes
````

````python
# Aqui alteramos as variáveis binárias object em 1 ou 0.

df_02.sexo.unique()
````

````python
df_02.posse_de_imovel.unique()
````

````python
df_02.posse_de_veiculo.unique()
````

````python
#Com o o método map substituo os dados para 1 e 0.
#Masculino 0, Feminino 1; 
#Tem imóvel 0, não tem 1;
#Tem veículo 0, não tem 1.

df_02.sexo = df_02.sexo.map({'M': 1, 'F':0})
df_02.posse_de_imovel = df_02.posse_de_imovel.map({'Y': 1, 'N': 0})
df_02.posse_de_veiculo = df_02.posse_de_veiculo.map({'Y': 1, 'N': 0})

df_02
````

````python
#Aqui vou transformar as variáveis com tipo boolean para int.

df_02.mau = df_02.mau.astype(int)

df_02.dtypes
````
