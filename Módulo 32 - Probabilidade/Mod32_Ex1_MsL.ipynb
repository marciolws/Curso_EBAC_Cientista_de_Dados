# Probabilidade - Tarefa

```python
#Importando as bibiliotecas a serem utilizadas.

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
def calcular_probabilidade_maior(media, desvio_padrao, limite_superior):
    """Calcula a probabilidade de uma variável aleatória normal se maior que um limite superior"""
    probabilidade = 1 - stats.norm.cdf(x=limite_superior, loc=media, scale=desvio_padrao)
    return probabilidade

def exibir_resultado(limite_superior, probabilidade):
    """Exibe o resultado da probabilidade formatado."""
    print(f'A probabilidade de um pacote mais que {limite_superior} kg é de: {probabilidade * 100:.2f}%'.replace('.', ','))

#Paramentros de distribuição normal.
media = 15
desvio_padrao = 0.1
limite_superior = 15.2

#Cálculo da probabilidade.
probabilidade = calcular_probabilidade_maior(media, desvio_padrao, limite_superior)

#Exibe o resultado
exibir_resultado(limite_superior, probabilidade)
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
#Parâmetros da distribuição normal.
media = 15
desvio_padrao = 0.1

#Calcula o percentil de ordem 95%.
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o pedo dos pacites é: {percentil_95:.2f} kg'.replace('.', ','))
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
# Valores de Y
Y = np.linspace(-6, 6, 200)

# Cálculo das probabilidades
Y1 = stats.norm.cdf(Y)
Y2 = stats.t.cdf(Y, df=5)

# Criação do DataFrame com os resultados
resultados = pd.DataFrame({
    'Y': [f'<= {x:.2f}' for x in Y],
    'Y1': [f'{x * 100:.2f}%' for x in Y1],
    'Y2': [f'{x * 100:.2f}%' for x in Y2]
})

# Exibição dos resultados
pd.set_option('display.max_rows', 200)
print(resultados)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
Y = np.linspace(-6, 6, 200)
Y1 = stats.norm.cdf(Y)
Y2 = stats.t.cdf(Y, df=5)

#Configurações do estido do gráfico.
sns.set(style='whitegrid')

#Criação da figura e do eixo.
plt.figure(figsize=(30, 9))

#Plotagem dos dados.
plt.plot(Y, Y1, label='Y1 - Normal')
plt.plot(Y, Y2, label='Y2 - t(5)')

#Configuração do gráfico.
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))
plt.legend()
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
#Criação do array de quantis.
P = np.linspace(0, 1, 20)

#Parâmetros para a distribuição normal padrão.
media = 0
desvio_padrao = 1 #Variância = 1 então o desvio padrão é 1

#Cálculos dos percentis.
Y1 = stats.norm.ppf(q=P, loc=media, scale=desvio_padrao)
Y2 = stats.t.ppf(q=P, df=5)

#Exibindo os resultados em um DF.
resultados = pd.DataFrame({
    'P': np.round(P, 2),
    'Y1': np.round(Y1, 2),
    'Y2': np.round(Y2, 2)
})

print(resultados)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Criação do array de percentis
P = np.linspace(0, 1, 20)

# Cálculo dos percentis
Y1 = stats.norm.ppf(q=P)
Y2 = stats.t.ppf(q=P, df=5)

# Criação do gráfico
plt.figure(figsize=(15, 5))  # Tamanho do gráfico
plt.plot(P, Y1, label='Y1 - Normal')
plt.plot(P, Y2, label='Y2 - t(5)')

# Configuração dos eixos e adição de rótulos
plt.xlabel('Percentis (P)')
plt.ylabel('Valores de Y1 e Y2')
plt.title('Comparação entre Y1 e Y2 em Função dos Percentis')
plt.legend()

# Exibição do gráfico
plt.grid(True)  # Adiciona uma grade ao gráfico
plt.show()
```
