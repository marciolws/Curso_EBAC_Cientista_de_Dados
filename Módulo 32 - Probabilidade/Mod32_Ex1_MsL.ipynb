# Probabilidade - Tarefa

```python
#Importando as bibiliotecas a serem utilizadas.

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

from scipy import stats
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
def calcular_probabilidade_maior(media, desvio_padrao, limite_superior):
    """Calcula a probabilidade de uma varivel aleatria normal se maior que um limite superior"""
    probabilidade = 1 - stats.norm.cdf(x=limite_superior, loc=media, scale=desvio_padrao)
    return probabilidade

def exibir_resultado(limite_superior, probabilidade):
    """Exibe o resultado da probabilidade formatado."""
    print(f'A probabilidade de um pacote mais que {limite_superior} kg  de: {probabilidade * 100:.2f}%'.replace('.', ','))

#Paramentros de distribuio normal.
media = 15
desvio_padrao = 0.1
limite_superior = 15.2

#Clculo da probabilidade.
probabilidade = calcular_probabilidade_maior(media, desvio_padrao, limite_superior)

#Exibe o resultado
exibir_resultado(limite_superior, probabilidade)
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
#Parmetros da distribuio normal.
media = 15
desvio_padrao = 0.1

#Calcula o percentil de ordem 95%.
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o pedo dos pacites : {percentil_95:.2f} kg'.replace('.', ','))
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Valores de Y
Y = np.linspace(-6, 6, 200)

# Clculo das probabilidades
Y1 = stats.norm.cdf(Y)
Y2 = stats.t.cdf(Y, df=5)

# Criao do DataFrame com os resultados
resultados = pd.DataFrame({
    'Y': [f'<= {x:.2f}' for x in Y],
    'Y1': [f'{x * 100:.2f}%' for x in Y1],
    'Y2': [f'{x * 100:.2f}%' for x in Y2]
})

# Exibio dos resultados
pd.set_option('display.max_rows', 200)
print(resultados)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
Y = np.linspace(-6, 6, 200)
Y1 = stats.norm.cdf(Y)
Y2 = stats.t.cdf(Y, df=5)

#Configuraes do estido do grfico.
sns.set(style='whitegrid')

#Criao da figura e do eixo.
plt.figure(figsize=(30, 9))

#Plotagem dos dados.
plt.plot(Y, Y1, label='Y1 - Normal')
plt.plot(Y, Y2, label='Y2 - t(5)')

#Configurao do grfico.
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade Acumulada')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))
plt.legend()
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
#Criao do array de quantis.
P = np.linspace(0, 1, 20)

#Parmetros para a distribuio normal padro.
media = 0
desvio_padrao = 1 #Varincia = 1 ento o desvio padro  1

#Clculos dos percentis.
Y1 = stats.norm.ppf(q=P, loc=media, scale=desvio_padrao)
Y2 = stats.t.ppf(q=P, df=5)

#Exibindo os resultados em um DF.
resultados = pd.DataFrame({
    'P': np.round(P, 2),
    'Y1': np.round(Y1, 2),
    'Y2': np.round(Y2, 2)
})

print(resultados)
```

6. Faa um grfico com os dados do item anterior.

```python
# Criao do array de percentis
P = np.linspace(0, 1, 20)

# Clculo dos percentis
Y1 = stats.norm.ppf(q=P)
Y2 = stats.t.ppf(q=P, df=5)

# Criao do grfico
plt.figure(figsize=(15, 5))  # Tamanho do grfico
plt.plot(P, Y1, label='Y1 - Normal')
plt.plot(P, Y2, label='Y2 - t(5)')

# Configurao dos eixos e adio de rtulos
plt.xlabel('Percentis (P)')
plt.ylabel('Valores de Y1 e Y2')
plt.title('Comparao entre Y1 e Y2 em Funo dos Percentis')
plt.legend()

# Exibio do grfico
plt.grid(True)  # Adiciona uma grade ao grfico
plt.show()
```
