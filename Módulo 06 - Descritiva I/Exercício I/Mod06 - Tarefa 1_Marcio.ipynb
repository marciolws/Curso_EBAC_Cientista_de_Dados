```python
#Antes de tudo, precisamos carregar as bibiliotes que serão utilizadas.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# %matplotlib notebook
```

```python
#Carregando o arquivo com pandas.
dados = pd.read_csv('sinasc_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#Aqui montamos o gráfico de barras com a frequencia para cada uma das variáveis.

plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(x='APGAR1', data=dados)
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(x='APGAR5', data=dados)
```

```python
#Gráfico de SEXO

plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(x='SEXO', data=dados)
```

```python
#Grafico gestação
plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(x='GESTACAO', data=dados)
```

```python
#Gráfico ESTCIVMAE

plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.countplot(x='ESTCIVMAE', data=dados)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
#Aqui montamos um gráfico de barras com o barplot usando o APGAR5 no eixo X e a CONSULTAS no eixo y.

plt.close('all')
plt.rc('figure', figsize=(10, 5))

sns.barplot(x='APGAR5', y='CONSULTAS', data=dados)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
valores_gestacao = dados['GESTACAO'].drop_duplicates().dropna().tolist()
```

```python
dados['APGAR1'].value_counts()
```

```python
valores_gestacao
```

```python
#Fechar os gráficos abertos
#Depois realizo o plot dos gráficos.

plt.close('all')

plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(nrows=2, ncols=2)

sns.countplot(ax=axes[0, 0], x='SEXO', data=dados)
sns.countplot(ax=axes[0, 1], x='GESTACAO', data=dados)
sns.countplot(ax=axes[1, 0], x='ESTCIVMAE', data=dados)
sns.countplot(ax=axes[1, 1], x='CONSULTAS', data=dados)

axes[0, 1].set_xticklabels(valores_gestacao, rotation=330)

plt.subplots_adjust(wspace=0.3, hspace=0.5)
```
