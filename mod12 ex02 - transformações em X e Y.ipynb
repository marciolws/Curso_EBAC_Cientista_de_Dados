### 1. Melhorando o ajuste no percentual de gorjetas

a. Vamos partir do modelo feito na última tarefa, o que relaciona ```tip_pct``` e ```net_bill```. Carregue a base, os pacotes e reproduza este modelo aqui.

````python
#Importando as bibliotecas que serão utilizadas.

import pandas as pd
import seaborn as sns
from seaborn import load_dataset

import matplotlib.pyplot as plt

import numpy as np

import statsmodels.formula.api as smf
````

````python
#Base de dados e calculando as variáveis que serão utilizadas.

tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
````

````python
tips.head()
````

````python
#Realizando a regressão linear do tip_pct e net_bill.tips.
#Notamos que o R² deu um valor bem baixo, cerca de 9%.
#Isso nos evidenciam que os dados não se ajustam a este modelo.

reg = smf.ols('tip_pct ~ net_bill', data=tips).fit()
reg.summary()
````

````python
#Plotando o gráfico da tip_pct e net_bill.

sns.regplot(y = 'tip_pct', x = 'net_bill', data=tips[tips['tip_pct']<.5])

x= np.linspace(0, tips['net_bill'].max(), 100)
y = -.006 * x

plt.plot(x, y, '-r', label='linha dos 10%')

plt.xlim(left=0)
plt.ylim(bottom=0)
````

### 2. ajuste o modelo de ```tip_pct``` em função do logaritmo de ```net_bill```.

````python
#Realizando o ajuste dos dados para função tip_pct em função do log de net_bill.

reg = smf.ols('tip_pct ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='tip_pct', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

###  3. ajuste o modelo de ```tip_pct``` em função de um polinômio de segundo grau em ```net_bill```.

````python
#Ajustando o modelo de tip_pct em função de um polinômio de segundo grau para o net_bill.

reg = smf.ols('tip_pct ~ net_bill + np.power(net_bill,2)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='tip_pct', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], reg.fittedvalues, 'r,--')
````

###  4. ajuste um modelo no log do ```tip_pct``` em função de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```

````python
#Ajustando o modelo do log de tip_pct em função de net_bill.

reg = smf.ols('np.log(tip_pct) ~ net_bill', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='tip_pct', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
#Com a utilização do log na variável resposta previsamos realizar uma operação antes de usar o reg.fittedvalues, precisamos calcular o exponencial do reg.fittedvalues, pois ela é o logaritmo da reg.

tips['pred_tip_pct'] = np.exp(reg.fittedvalues)
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')
````

````python
#Para calcular o R² usamos a correlação entre o valor predito (reg.fittedvalues) e a variável resposta (tip_pct). O valor da correlação é R, então precisamos elevalo ao quadrado para obtermos R².

tips['pred'] = reg.fittedvalues

r = tips[['pred', 'tip_pct']].corr().iloc[0,1]
````

###  5. ajuste um modelo no log do ```tip_pct``` em função do log de ```net_bill```. Calcule o $R^2$ também em ```tip_pct```.

````python
#Aqui realizamos o log de ambas as variáves, resposta e explicativa.

reg = smf.ols('np.log(tip_pct) ~ np.log(net_bill)', data=tips).fit()
print(reg.summary())
tips['res'] = reg.resid

sns.scatterplot(x='net_bill', y='tip_pct', data=tips, alpha=.75)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
sns.regplot(x='net_bill', y='tip_pct', data=tips)
plt.plot(tips['net_bill'], tips['pred_tip_pct'], 'r,--')
````
