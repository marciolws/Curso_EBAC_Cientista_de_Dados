```python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
#Aqui vizualizamos o CSV
sinasc.head()
```

```python
sinasc.info()
```

```python
#A função isna cria uma dataframe com True e False. True para os dados faltantes e False para os dados não faltantes.
#Se realizarmos a soma nas colunas iremos obter o número faltantes de dado onde True é e False, zero.

sinasc.isna().sum()
```

```python
#Criado uma lista com as colunas que são relevantes para a análise.

lista_colunas = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO',
                 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
```

```python
#Usando a funcao loc para pegar apenas as colunas desejadas.
df_sinasc = sinasc.loc[:, lista_colunas]
```

```python
df_sinasc
```

```python
df_sinasc.isna().sum()
```

```python
#Para remover as linhas com APGAR5 = NaN utiliza-se a função isna()

df_sinasc_apgar5 = df_sinasc[df_sinasc['APGAR5'].isna()]    
```

```python
#Depois pegamos o index dessas colunas e as removemos do dataframe.
df_sinasc_clear = df_sinasc.drop(index=df_sinasc_apgar5.index)
```

```python
df_sinasc_clear.isna().sum()
```

```python
#A função fillna, na coluna ESTICIVMAE, substituimos os dados faltantes por 9.
#O mesmo poderia ser feito na coluna CONSULTAS, mas não há valores faltantes nela.

df_sinasc_clear['ESTCIVMAE'].fillna(value=9, inplace=True)
```

```python
df_sinasc_clear.isna().sum()
```

```python
#Realizei o mesmo processo na coluna QTDFILVIVO, substituindo os dados vazios por zero.

df_sinasc_clear['QTDFILVIVO'].fillna(value=0, inplace=True)
```

```python
df_sinasc_clear.isna().sum()
```

```python
#O demais dados faltantes foram removidas do dataframe, pois constatou-se que os dados faltantes representam uma porcentagem muito pequena do dataframe.

df_sinasc_clear.shape
```

```python
df_sinasc_clear.isna().sum()/26925
```

```python
df_sinasc_clear.dropna(inplace=True)
```

```python
df_sinasc_clear.isna().sum()
```

```python
df_sinasc_clear.shape
```

```python
#Usando a caracterização separamos o APGAR5 em 4 grupos.

df_sinasc_clear.loc[df_sinasc_clear['APGAR5'] <= 3, 'APGAR_CAT'] = 'asfixia severa'
df_sinasc_clear.loc[(df_sinasc_clear['APGAR5'] >= 4) & (df_sinasc_clear['APGAR5'] <= 5), 'APGAR_CAT'] = 'asfixia moderada'
df_sinasc_clear.loc[(df_sinasc_clear['APGAR5'] >= 6) & (df_sinasc_clear['APGAR5'] <= 7), 'APGAR_CAT'] = 'asfixia leve'
df_sinasc_clear.loc[df_sinasc_clear['APGAR5'] >= 8 , 'APGAR_CAT'] = 'normal'
```

Entre 8 e 10 está em uma faixa 'normal'.

Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.

Entre 4 e 5 significa 'asfixia moderada'.

Entre 0 e 3 significa 'asfixia severa'.
