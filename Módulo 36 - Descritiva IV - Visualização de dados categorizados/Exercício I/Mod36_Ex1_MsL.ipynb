## Anlise da probabilidade de retorno

Considere somente os dados de janeiro de 2020 em diante para esta anlise.

## Clculo do WOE

- Calcule o WOE das categorias, sendo que o evento em questo  a nota ter retorno>0.
- Analise este WOE por tabela e por grfico.

````python
# Tabela cruzada
tab = pd.crosstab(index=df_jan["categoria"], columns=df_jan["flag_credito"], margins=True, margins_name="Total") 

# Exibe a tabela cruzada
tab  
````

````python
 # Proporo de no evento
tab["perc_nao_evento"] = tab[0] / tab.loc["Total", 0] 

# Proporo de evento
tab["perc_evento"] = tab[1] / tab.loc["Total", 1]  

# Weight of Evidence (WOE)
tab["WOE"] = np.log(tab["perc_evento"] / tab["perc_nao_evento"])  

# Exibe a tabela com as novas colunas
tab  
````

````python
# Grfico de barras para a coluna WOE
tab["WOE"].plot.bar()  

# Linha horizontal pontilhada
plt.axhline(linestyle='--')  

 # Rotao dos rtulos do eixo x
plt.xticks(rotation=45) 

# Exibe o grfico
plt.show()  
````

## Clculo do *Information Value*

Calcule o *IV* dessa varivel.

````python
# Calcula o IV parcial
tab["IV_parcial"] = (tab["perc_evento"] - tab["perc_nao_evento"]) * tab["WOE"]  

# Exibe o IV total
print(f"Information Value da varivel 'categoria': {tab['IV_parcial'].sum():.2%}\n") 

# Exibe a tabela com a nova coluna
tab 
````
