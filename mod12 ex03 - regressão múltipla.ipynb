### I. Modelo no valor da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo em ```tip```, explicada por ```sex, smoker, diner e net_bill```.  
2. Remova as variáveis não significantes.  
3. observe o gráfico de resíduos em função de ```net_bill```  
4. teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.

````python
#Importando as bibliotecas a serem usadas.

import pandas as pd
import numpy as np
from seaborn import load_dataset
import seaborn as sns
import matplotlib.pyplot as plt

import patsy
import statsmodels.api as sm
import statsmodels.formula.api as smf
````

````python
#Carregando a base de dados.

tips = load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
````

````python
tips.head()
````

````python
#Aqui vou usar o patsy para obter as matrizes de X e y.

y, X = patsy.dmatrices('tip ~ sex + smoker + time + net_bill + 0', data=tips)
````

````python
X
````

````python
y
````

````python
#Usando o statsmodel para fazer a modelagem dos dados.

modelo = sm.OLS(y, X).fit()
````

````python
#Aqui utilizo o summary para ver a modelagem.

modelo.summary()
````

````python
#Removendo os dados que apresenta os piores p-value.
#Após remover dados, nada mudou no valor de R², reforçando a hipótese de que sex não tem significancia para modelagem.

y, X = patsy.dmatrices('tip ~ sex + smoker + net_bill + 0', data=tips)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

````python
#Removendo a variável smoker, diminuiu ligeiramente o valor de R².

y, X = patsy.dmatrices('tip ~ sex + net_bill + 0', data=tips)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

````python
#Plotando o gráfico de resíduos.

tips['res'] = modelo.resid

sns.scatterplot(x='net_bill', y='res', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
#Modelagem de dados com log de net_bill
#O net_bill mantem a significância estatística das variáveis, prém, apresenta um R² menor.

y, X = patsy.dmatrices('tip ~ smoker + np.log(net_bill) + 0', data=tips)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

````python
#Plotando o resíduo, em escala logaritmica, está menos disperso.

tips['log_net_bill'] = np.log(tips['net_bill'])

tips['res_log'] = modelo.resid

sns.scatterplot(x='log_net_bill', y='res_log', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
#Dados com log de net_bill. O valor de R² ficou entre o net_bill e log(net_bill)

y, X = patsy.dmatrices('tip ~ smoker + np.power(net_bill, 2) + 0', data=tips)

modelo = sm.OLS(y, X).fit()
modelo.summary()
````

````python
#Plotando o resíduo, o net_bill como um polinômio de segundo grau ocorreu novamente uma dispersão maior do resíduo.

tips['pol2_net_bill'] = np.power(tips['net_bill'], 2)

tips['res_pol2'] = modelo.resid

sns.scatterplot(x='pol2_net_bill', y='res_pol2', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

### Observação:

Levando-se em consideração o valor de R², o modelo que descreve melhor os dados é o net_bill.

### II. Modelo no valor do percentual da gorjeta

1. Crie a matriz de design (e a matriz y) utilizando o Patsy, para um modelo no log de ```tip```, explicado por ```sex, smoker, diner e net_bill```.
2. Remova as variáveis não significantes.
3. Observe o gráfico de resíduos em função de ```net_bill```
4. Teste transformar ```net_bill``` no log e um polinômio. Escolha o melhor modelo.
5. Do modelo final deste item, calcule o $R^2$ na escala de ```tip``` (sem o log). Compare com o modelo do item 1. Qual tem melhor coeficiente de determinação?

````python
#A modelagem do log de tip, vamos realizar os mesmos passos descritos acima.

tips = load_dataset('tips')
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips['net_bill'] = tips['total_bill'] - tips['tip']
````

````python
#Modelagem dos dados

y, X = patsy.dmatrices('np.log(tip) ~ sex + smoker + time + net_bill + 0', data=tips)
````

````python
#Utilizando o stasmodels obtemos as estatísticas da modelagem.

modelo_1 = sm.OLS(y, X).fit()
modelo_1.summary()
````

````python
#Removendo time da modelagem. A remoção da variável não alterou o valor de R².

y, X = patsy.dmatrices('np.log(tip) ~ sex + smoker + net_bill + 0', data=tips)

modelo_1 = sm.OLS(y, X).fit()
modelo_1.summary()
````

````python
#Removendo time da modelagem. A remoção da variável diminuiu levemente o valor de R²

y, X = patsy.dmatrices('np.log(tip) ~ sex + net_bill + 0', data=tips)

modelo_1 = sm.OLS(y, X).fit()
modelo_1.summary()
````

````python
#Realizando uma operação antes de usar fittedvalues

tips['pred_net_bill'] = np.exp(modelo_1.fittedvalues)
````

````python
#Plotando gráfico de dispersão do residuo.

tips['res'] = modelo_1.resid

sns.scatterplot(x='pred_net_bill', y='res', data = tips)
plt.axhline(y=0, color='r', linestyle='--')
````

````python
#Modelagem com log de net_bill. O valor de R² aumentou e as variáveis continuam apresentando significância estatística.

y, X = patsy.dmatrices('np.log(tip) ~ sex + np.log(net_bill) + 0', data=tips)

modelo_1 = sm.OLS(y, X).fit()
modelo_1.summary()
````

````python
#Modelagem com polinômio de segundo grau. O valor de R² diminui consideravelmente comparado aos demais modelos.

y, X = patsy.dmatrices('np.log(tip) ~ sex + np.power(net_bill, 2) + 0', data=tips)

modelo_1 = sm.OLS(y, X).fit()
modelo_1.summary()
````

### Observações:

O melhor modelo, que apresentou o maior valor de R² e manteve a significância estatística das variáveis é o log de net_bill.

Comparando o melhor modelo de tip e log tip, log tip apresentou o melhor valor de R² explicando melhor os dados.
