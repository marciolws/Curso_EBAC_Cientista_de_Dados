#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na tempreatura global, A NASA disponibiliza dados de diferenças de de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|
| 1900 | -0.08 |
| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

```python
import pandas as pd
import numpy as np

# Seu código aqui


dados_temp = [-0.08, -0.27, 0.12, -0.03, 0.26, 0.40, 1.02] #Lista com dados das anomalias termicas
dados_ano = [1900, 1920, 1940, 1960, 1980, 2000, 2020] #Lista com os anos que aconteceram as anomalias

df = pd.Series(dados_temp) #Aqui foi transformado a lista de temperatura em uma serie usando o pandas.
```

```python
df #Esse codigo imprimo a serie.
```

#### 2) Coloque os anos nos índices conforme a tabela.

```python

# seu código aqui
df = pd.Series(dados_temp, index=dados_ano) #Colocando no parâmetro 'index' a coluna dos anos para torna-las o index da série.
```

#### 4) Transforme o ndarray abaixo em um dataframe. 
O numpy é capaz de gerar arrays n-dimensionais com números pseudo-aleatórios de acordo com uma variedade de distribuições, como no exemplo abaixo. Transforme esse nd-array em um DataFrame.

```python
arr = np.random.normal(100, 10, (20,3))

# seu código aqui
arr #É uma array formada aleatoriamente, com 20 linhas e 03 colunas.
```

```python
df = pd.DataFrame(data=arr) #Transformamos a ndarray em um data frame
```

```python
df
```

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#seu código aqui
df.columns = ['x1', 'x2', 'x3'] #Aplicando o método columns ao dataframe podemos indica.
```

```python
df
```

```python
#Para colocar o indice de 1 a 20 foi criada uma lista vazia e colocado nela os valores desejados a partir de um laço de repetição.
df_index = []
i = 1

while i < 21:
    df_index.append(i)
    i += 1

# Usado o método index para inserir o index desejado ao dataframe.
df.index = [df_index]
```

```python
df
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui

#Foi criada uma coluna média e inserido o valor médio das 03 colunas de dados.
df['media'] = (df['x1'] + df['x2'] + df['x3']) / 3
```

```python
df
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui

# Criada a coluna log_media e inserido o logaritmo natural ou neperiano através do numpy usando o método log1p.
df['log_media'] = np.log1p(df['media'])
```
