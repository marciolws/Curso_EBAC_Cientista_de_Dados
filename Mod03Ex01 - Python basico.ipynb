### 04 - Números primos
> Um número **N** é primo se e somente se é divisível por 1, -1, por **N** e por -**N**.  

Escreva um script que verifica se ```N``` é um número primo, verificando se ```N``` é divisível por todos os números de ```1``` a ```N-1```. Você vai precisar usar alguma ferramenta de *loop* que você aprendeu para isto. No final, devolva uma mensagem no output indicando se o número é primo ou não.

````python
N = 47

# seu código abaixo

contador = 2 # O contador é inicia no 2 pois todos os números são divisíveis por 1.
primo = 0 # Essa variável indicara se um número é primo ou não. Se ao final da verificação ele continuar sendo igual à 0, então o número é primo.

if N != 1 and N != -1 and N != 0: # O número 1 e -1 não são primos, não podem dividir por zero.

    if N != 2 and N != -2: # O número 2 é primo, o único número par que é primo. Daria para fazer sem esse if mas teria que colocar outro if perguntado se primo é igual à 1 e N é igual a 2.

        while contador < N: # O while irá fazer a repetição até um N-1.
            if N % contador == 0: # Se o resto da divisão for igual à 0, então ele não é primo.
                primo += 1
            contador += 1
        if primo == 0: # Terminada a repetição, verifica se o valor de primo continua sendo 0.
            print(f'{N} é um número primo.')
        else:
            print(f'{N} não é um número primo.')
            
    else:
        print('Dois é primo.')
else:
    print('Os números "0", "1" e "-1" não são primos.')
````

2. Será que precisamos mesmo verificar **todos** os números?
3. Será que precisamos ir até N-1?

Essas perguntas levam ao tipo de pensamento voltado a deixar um algoritmo mais eficiente. Veja se você consegue melhorar o seu.

````python
N = 98

# seu código aqui

'''O loop não necessariamente precisa ir até o final sempre, ele pode ir até o "N/2" ou algo perto, talvez N/2+1 ou N/2-1.
Ele pode verificar apenas os números ímpares, dado que o único número primo que é par é o número 2.
Se o número logo no início já for divisível, posso parar a repetição.'''

contador = 3 #Inicia no 3 pois 2 é o único número par que é primo.
primo = 0

if N == 2 or N == -2: #Faz a verificação do 2.
    print('Dois é primo.')
elif N == 1 or N == -1 or N == 0: #Realiza a verificação do 1, -1 e 0.
    print('"0", "1" e "-1" não são primos.')
elif N % 2 == 0: #Verifica os números pares.
    print(f'{N} não é primo')
else:
    while contador < N/2: #Essa repetição vai somente até a metade do número.
        if N % contador == 0:
            primo += 1

        if primo != 0:
            break
        
        contador += 2 #Verifico apenas os números ímpares.
    
    if primo == 0: #Verificação final para saber se um número é primo ou não.
        print(f'{N} é primo.')
    else:
        print(f'{N} não é primo.')
````

### 07 - Peso ideal 2
Recebendo um valor de altura, encontre o peso '*ideal*' dessa pessoa, que fornece o IMC encontrado acima

````python
altura = 1.70

# Seu código

imc_ideal = 21.7 # O IMC ideal encontrado no exercício anterior foi de 21.7

# Utilizando a equação fornecida foi calculado o peso ideal desta pessoa para a altuda informada.
peso_ideal = imc_ideal * (altura ** 2)

peso_ideal = round(peso_ideal, ndigits= 2)

print(peso_ideal)
````

### 08 - Peso ideal 3
Dada uma lista contendo as alturas de pacientes, crie uma nova lista que contenha o peso '*ideal*' (que fornece o IMC calculado em **Peso ideal 1**) desses pacientes.

````python
lista_alturas = [1.95, 2.05, 1.70, 1.65]

lista_peso_ideal = []

# seu código

imc_ideal = 21.7 #Inserimos novamente o valor do imc ideal.

for altura in lista_alturas: #Realizado o cálculo para cada altura usando um laço de repetição.
    peso_ideal_1 = imc_ideal * (altura ** 2)
    lista_peso_ideal.append(round(peso_ideal_1, ndigits=2))

print(lista_peso_ideal)
````

### 09 - Peso ideal 4
Dada uma lista de tuplas - cada elemento da lista é uma tupla contendo altura e peso de um paciente - crie uma nova lista com o IMC desses pacientes.

````python
altura_peso = [(1.80, 90), (1.65, 75), (1.91, 70)]

imc = []

# seu código

for dado in altura_peso: #Foi distribuido o valor da altura e peso para variáveis.
    altura, peso = dado 
    imc_calculado = peso / (altura ** 2) #Calculo do imc.
    imc.append(round(imc_calculado, ndigits=2))  #Adicionando a lista.
print(imc)
````
